```python
import pandas as pd
```

```python
pred = pd.read_csv("../datasets/covid/infer_rslt.csv")
pred.drop(columns={"Unnamed: 0"}, inplace=True)
pred
```

```python
for i in range(4):
    irrelavant = pred[pred["y_true"] == i]
    irrelavant_correct = irrelavant[irrelavant["y_pred"]==i]
    print(i,len(irrelavant_correct)/len(irrelavant))
```

```python
pred[pred["y_true"] == 0]
```

```python
map=dict(zip([i for i in range(4)],['positive', 'neutral', 'negative', 'unrelated']))
map
```

```python
pred["y_true"] = pred["y_true"].apply(lambda x: map[x])
pred["y_pred"] = pred["y_pred"].apply(lambda x: map[x])
pred
```

```python
test = pd.read_csv("../datasets/covid/test_NLI_M.csv",header=None,sep="\t")
test.columns = ["id", "y_true", "category", "text"]
test
```

```python
result = test.merge(pred, on = "y_true")
result
```

# Aspect Detection

```python
map1=dict(zip(['positive', 'neutral', 'negative', 'unrelated'], [0, 0, 0, 1]))
map1
```

```python
detection = result.copy()
detection["y_true"] = detection["y_true"].apply(lambda x: map1[x])
detection["y_pred"] = detection["y_pred"].apply(lambda x: map1[x])
detection
```

```python
from sklearn.metrics import f1_score
print("The Aspect Detection Task:")
for i in detection.category.unique():
    tmp = detection[detection["category"]==i]
    macrof1 = f1_score(tmp["y_true"], tmp["y_pred"], average='macro')
    microf1 = f1_score(tmp["y_true"], tmp["y_pred"], average='micro')
    print(f"Category: {i}\tMacro F1: {macrof1}\tMicro F1: {microf1}")
```

# Sentiment Classification

```python
classification = result.copy()
classification = classification[classification["y_true"] != "unrelated"]
classification.reset_index(drop=True, inplace=True)
classification
```

```python
classification["score"] = classification["score"].apply(lambda x: [ float(i) for i in x[1:-1].split(" ") if len(i)>0 ][:-1])
classification["y_pred"] = classification["score"].apply(lambda x: x.index(max(x)))
map2=dict(zip(['positive', 'neutral', 'negative'], [0, 1, 2]))
classification["y_true"] = classification["y_true"].apply(lambda x: map2[x])
classification[classification["y_pred"] == "unrelated"]
```

```python
classification
```

```python
print("The Sentiment Classification Task:")
for i in classification.category.unique():
    tmp = classification[classification["category"]==i]
    macrof1 = f1_score(tmp["y_true"], tmp["y_pred"], average='macro')
    microf1 = f1_score(tmp["y_true"], tmp["y_pred"], average='micro')
    print(f"Category: {i}\tMacro F1: {macrof1}\tMicro F1: {microf1}")
```

# Joint Task

```python
print("The Joint Task:")
for i in result.category.unique():
    tmp = result[result["category"]==i]
    macrof1 = f1_score(tmp["y_true"], tmp["y_pred"], average='macro')
    microf1 = f1_score(tmp["y_true"], tmp["y_pred"], average='micro')
    print(f"Category: {i}\tMacro F1: {macrof1}\tMicro F1: {microf1}")
```
